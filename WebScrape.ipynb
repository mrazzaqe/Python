```python
from bs4 import BeautifulSoup
import requests
import re

# Downloading imdb top 250 movie's data
url = "http://www.imdb.com/chart/top"
req = requests.get(url)
soup = BeautifulSoup(req.text, "lxml")
movies = soup.select('td.titleColumn')
links = [a.attrs.get('href') for a in soup.select('td.titleColumn a')]
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value') for b in soup.select('td.posterColumn span[name=ir]')]
votes = [b.attrs.get('data-value') for b in soup.select('td.ratingColumn strong')]
```

```python
# create a empty list for storing
# movie information
list = []

# Iterating over movies to extract
# each movie's details
for index in range(0, len(movies)):
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    place = movie[:len(str(index))-(len(movie))]
    data = {"movie_title": movie_title,
            "year": year,
            "place": place,
            "star_cast": crew[index],
            "rating": ratings[index],
            "vote": votes[index],
            "link": links[index]}
    list.append(data)
```

```python
for movie in list:
    print(movie['place'], '-', movie['movie_title'], '-', movie['year'],
          '-', movie['star_cast'], '-',movie['rating'])
```

```python
#Searching and navigating through the parse tree
from bs4 import BeautifulSoup
import requests
url = "https://www.daraz.com.bd/smartphones/"
req = requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser')

smartphones =[]

table = soup.find('div', attrs = {'id':'smartphones'}) 

for row in table.findAll('div',
                         attrs = {'class':'col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top'}):
    smartphones = {}
    smartphones['theme'] = row.h5.text
    smartphones['url'] = row.a['href']
    smartphones['img'] = row.img['src']
    smartphones['lines'] = row.img['alt'].split(" #")[0]
    smartphones['author'] = row.img['alt'].split(" #")[1]
    smartphones.append(smartphones)
   
filename = 'smartphones.csv'
with open(filename, 'w', newline='') as f:
    w = csv.DictWriter(f,['theme','url','img','lines','author'])
    w.writeheader()
    for quote in smartphones:
        w.writerow(smartphones)
```

```python
#Python program to scrape website 
#and save quotes from website
import requests
from bs4 import BeautifulSoup
import csv
   
URL = "http://www.values.com/inspirational-quotes"
r = requests.get(URL)
   
soup = BeautifulSoup(r.content, 'html5lib')
   
quotes=[]  # a list to store quotes
   
table = soup.find('div', attrs = {'id':'all_quotes'}) 
   
for row in table.findAll('div',
                         attrs = {'class':'col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top'}):
    quote = {}
    quote['theme'] = row.h5.text
    quote['url'] = row.a['href']
    quote['img'] = row.img['src']
    quote['lines'] = row.img['alt'].split(" #")[0]
    quote['author'] = row.img['alt'].split(" #")[1]
    quotes.append(quote)
   
filename = 'inspirational_quotes.csv'
with open(filename, 'w', newline='') as f:
    w = csv.DictWriter(f,['theme','url','img','lines','author'])
    w.writeheader()
    for quote in quotes:
        w.writerow(quote)
```
